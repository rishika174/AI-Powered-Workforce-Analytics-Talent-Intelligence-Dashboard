Average Marks by CourseThe average marks scored by students vary across different academic branches, with Electronics recording the highest performance:CourseAverage MarksElectronics92.0%Computer Science86.5%Civil81.0%Mechanical78.0%

```python
import pandas as pd

# Create sample student data
data = {
    "StudentID": [201, 202, 203, 204, 205],
    "Name": ["Riya", "Aman", "Neha", "Karan", "Priya"],
    "Age": [20, 21, 19, 22, 20],
    "Course": ["Computer Science", "Mechanical", "Electronics", "Computer Science", "Civil"],
    "Marks": [85, 78, 92, 88, 81]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display data
print(df)
```

```python

print("\nTotal Students:", len(df))
```

```python
print("Average Age:", df["Age"].mean())
```

```python
print("Average Marks:", df["Marks"].mean())
```
