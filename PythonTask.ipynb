```python
import pandas as pd

# Create sample student data
data = {
    "StudentID": [201, 202, 203, 204, 205],
    "Name": ["Riya", "Aman", "Neha", "Karan", "Priya"],
    "Age": [20, 21, 19, 22, 20],
    "Course": ["Computer Science", "Mechanical", "Electronics", "Computer Science", "Civil"],
    "Marks": [85, 78, 92, 88, 81]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display data
print(df)

print("\nTotal Students:", len(df))
print("Average Age:", df["Age"].mean())
print("Average Marks:", df["Marks"].mean())

print("\nStudents in each Course:")
print(df["Course"].value_counts())
```
