### SQL Database Operations on Student Records

This notebook demonstrates basic CRUD operations and data analysis using **SQLite** and **pandas** in Python:

*   **Database Setup:** Creates a `Students` table with fields for ID, Name, Course, Marks, and Semester.
*   **Data Insertion:** Populates the table with sample student records.
*   **Data Querying & Aggregation:**
    *   Filters students with marks greater than 80.
    *   Calculates the average marks grouped by each course.
    *   Filters students enrolled in semesters higher than 4.
    *   Identifies the top-performing student with the highest marks.
*   **Data Cleanup:** Demonstrates how to delete a specific record using a `WHERE` clause.

```python
import sqlite3
import pandas as pd

# Create database
conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Drop table if it already exists
cursor.execute("DROP TABLE IF EXISTS Students")

# Create Students table
cursor.execute("""
CREATE TABLE Students (
    StudentID INTEGER PRIMARY KEY,
    Name TEXT,
    Course TEXT,
    Marks INTEGER,
    Semester INTEGER
)
""")

# Insert sample student data
students = [
    (1, "Riya", "Computer Science", 85, 3),
    (2, "Aman", "Mechanical", 78, 4),
    (3, "Neha", "Computer Science", 92, 5),
    (4, "Karan", "Electronics", 88, 6),
    (5, "Priya", "Civil", 81, 3),
    (6, "Arjun", "Mechanical", 75, 5),
    (7, "Sneha", "Computer Science", 95, 7),
    (8, "Vikram", "Civil", 83, 4)
]
```

```python
cursor.executemany("INSERT INTO Students VALUES (?, ?, ?, ?, ?)", students)
conn.commit()

print("\nAll Students Data")
display(pd.read_sql("SELECT * FROM Students", conn))
```

```python

print("\nStudents with Marks > 80")
display(pd.read_sql("""
SELECT Name, Marks
FROM Students
WHERE Marks > 80
""", conn))
```

```python
print("\nAverage Marks by Course")
display(pd.read_sql("""
SELECT Course,
AVG(Marks) AS Average_Marks
FROM Students
GROUP BY Course
""", conn))
```

```python
print("\nStudents in Semester > 4")
display(pd.read_sql("""
SELECT *
FROM Students
WHERE Semester > 4
""", conn))
```

```python
print("\nStudent with the Highest Marks")
display(pd.read_sql("""
SELECT *
FROM Students
ORDER BY Marks DESC
LIMIT 1
""", conn))
```

```python

# Delete a student record
cursor.execute("""
DELETE FROM Students
WHERE StudentID = 5
""")

conn.commit()
conn.close()
```
